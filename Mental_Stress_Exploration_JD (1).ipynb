```python


import zipfile
import os

zip_path = "/content/archive (1).zip"  # <- upload your zip again in Colab
extract_path = "/content/mental_stress_dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# See what files are inside
for root, dirs, files in os.walk(extract_path):
    for file in files:
        print(os.path.join(root, file))
```

```python

import pandas as pd

df = pd.read_csv("/content/mental_stress_dataset/Student_Mental_Stress_and_Coping_Mechanisms.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Mental Stress Level', y='Academic Performance (GPA)', data=df)
plt.title("📚 Stress vs GPA")
plt.show()
```

```python
sns.boxplot(x='Mental Stress Level', y='Sleep Duration (Hours per night)', data=df)
plt.title("😴 Sleep vs Stress")
plt.show()
```

```python
df['Stress Coping Mechanisms'].value_counts().head(10).plot(kind='barh', color='orange')
plt.title("🧘‍♂️ Top Coping Mechanisms")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("📊 Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Gender', y='Mental Stress Level', data=df, palette='Set2')
plt.title("🚻 Gender vs Stress")
plt.show()
```
